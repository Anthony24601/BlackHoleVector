```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
fileList = [f for f in os.listdir("../EHT_M87") if f.endswith('.txt')]  #open that folder 
fileList
```

```python
for f in fileList:
    file = open(f"../EHT_M87/{f}","r")
    lines = file.readlines()
    lines = lines[9:]
    image = np.empty([180,180])
    t = []
    for line in lines:
        coord = line.strip().split()
        x = int(float(coord[0]) * 1000000)
        y = int(float(coord[1]) * 1000000)
        z = int(float(coord[2]) * 10000000000)
        image[x+90][y+90] = z
    plt.figure()
    plt.imshow(image)
    plt.axis('off')
    plt.savefig(f'../images/{f.split(".")[0]}.png')
```
